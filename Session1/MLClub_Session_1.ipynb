# Intro to Decorators

Python 'decorator' is a very powerful syntax, actually a function or a class that wraps another function or a method. It is said that the decorator 'decorates' the other function or method thereby replacing it; i'll explain how. 

Decorator can be used in code by using the @decorator syntax and provides a wide range of functionalities which a programmer can capitalize. Let me give you an example.

### My first decorator

```python
class myDecorator(object):

    def __init__(self, f):
        print "inside myDecorator.__init__()"
        f() # Prove that function definition has completed

    def __call__(self):
        print "inside myDecorator.__call__()"

@myDecorator
def aFunction():
    print "inside aFunction()"

print "Finished decorating aFunction()"

aFunction()
```

### Decorator order when decorating a class

```python
class entryExit(object):

    def __init__(self, f):
        self.f = f

    def __call__(self):
        print "Entering", self.f.__name__
        self.f()
        print "Exited", self.f.__name__

@entryExit
def func1():
    print "inside func1()"

@entryExit
def func2():
    print "inside func2()"

func1()
func2()
```

### Decorator for decorating functions

```python
# Decorator function
def entryExit(f):
    # new_f has access to `f` since it is within entryExit decorator.
    def new_f():
        print "Entering", f.__name__
        f()
        print "Exited", f.__name__
    return new_f

@entryExit
def func1():
    print "inside func1()"

@entryExit
def func2():
    print "inside func2()"

func1()
func2()
```

```python
# Basically, this is what is happening !
# You call the function inside the decorator.
entryExit(func1)()
```

## Use Case for Python Decorators

Most of the time when running some compute intensive code, I often find the execution time for small parts of the code using Python's `time` module. For example, by logging time in the beginning and end of a function definition, I can find the execution time in seconds to run the function. The seconds can be formatted in to hours, minutes and seconds (H:M:S) later. The problem is that I have to manually write the code for finding execution time for each function, and that is sad.

```python
"""
This is how I used to find execution time in Python.
"""
import time
from datetime import timedelta
import numpy as np

def run_code():
    """ Function to run the training """
    # Log the start time
    start_time = time.time()
    # Call the compute intense function (Dummy code shown here.)
    
    a = np.ones(1000000)
    b = np.ones(1000000)
    c = np.multiply(a,b)
    
    # Log the end time
    end_time = time.time()
    # Print the execution time
    print (("Time taken to execute the function is {ex_time}."
       ).format(ex_time=str(timedelta(seconds=end_time-start_time))))

# Run the function with extra code for timing.
run_code()
# Here timedelta converts seconds to H:M:S.
```

Decorators show its versatility and reproducibility in these situations. Instead of logging time before and after the compute intense task inside the function, I can write a decorator to find the execution time and call the compute intense function inside it. Here, the decorator will decorate or wrap the compute intense function inside it.

```python
"""
This is how I do timing with Python Decorators. A full working example.
"""
import time
from datetime import timedelta

# Define the decorator here.

def timeToRun(f):
    """The decorator function."""
    def findTime():
        """The actual function which is used to 
        decorate the passed functions."""
        # Log the start time
        st = time.time()
        # Call the passed function 
        f()
        # Log the end time
        ed = time.time()
        print (("Time taken to execute the function {f_name} is {time:}").format
               (time=str(timedelta(seconds=ed-st)),f_name=f.__name__))
    # Return the decorated function
    return findTime


@timeToRun
def test_time_to_run():
    """Dummy compute function"""
    a = np.ones(1000000)
    b = np.ones(1000000)
    c = np.multiply(a,b)

# Run the decorated function
test_time_to_run()
```

The above snippet show's how easy it is to use a decorator. Here are some key points:

- By specifying `@timeToRun` just above the function definition of `test_time_to_run`, the latter function gets decorated by the former.

- Essentially, the function now becomes:

- Since a function can be decorated using @decorator syntax, it can be used to decorate other functions as well. This results in much cleaner and shorter code.

- The internal name of decorated function changes to the decorator:

```python
"""
Here test_time_to_run becomes:

    <function __main__.findTime>

Also, running:

    test_time_to_run.__name__   

gives:

    'findTime'


To keep the old function name, use the `wraps`
function from functools library inside the decorator.

For example, in the decorator:
"""

from functools import wraps
def timeToRun(f):
    # Add the wraps function
    @wraps(f)
    def findTime():
        st = time.time()
        f()
        ed = time.time()
        print (("Time taken to execute the function {f_name} is {time:}").format
               (time=str(timedelta(seconds=ed-st)),f_name=f.__name__))
    return findTime

@timeToRun
def test_time_to_run():
    """Dummy compute function"""
    a = np.ones(1000000)
    b = np.ones(1000000)
    c = np.multiply(a,b)
```
